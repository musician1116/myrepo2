```python
# 행렬 만들기,, 수학 연산, 행렬 연산(선형대수)...
import numpy as np
```

```python
# 시각화 라이브러리
#   matplotlib
#   seaborn
```

```python
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,14,100) # linspace(시작값, 끝값, 개수)
x
```

```python
y = np.sin(x)
plt.plot(x, y)
```

```python
y1 = np.sin(x)
y2 = np.sin(x) * 2
y3 = np.sin(x) * 4
plt.plot(x, y1, x, y2, x, y3)
```

```python
y1 = np.sin(x)
y2 = np.sin(x) * 2
y3 = np.sin(x) * 4
plt.plot(x, y1,label='sin(x)')
plt.plot(x, y2,label='sin(x) * 2')
plt.plot(x, y3, label='sin(x) * 4')
plt.legend()
plt.title('sin graph')
```

```python
y1 = np.cos(x)
y2 = np.cos(x) * 2
y3 = np.cos(x) * 4
plt.plot(x, y1, 'o', label='sin(x)')
plt.plot(x, y2, 'v', label='sin(x) * 2')
plt.plot(x, y3, '^', label='sin(x) * 4')
plt.legend()
plt.title('sin plot')
```

```python
# [데이터].info() : 데이터 정보
# [데이터].describe() : 수치형 데이터 요약값
# [데이터].shape : 행, 열의 정보
# [데이터].columns : 컬럼명, 제목
# [데이터].head() : 앞에서 기본적인 데이터 몇개(5)를 확인해 준다.
# [데이터].tail() : 뒤에서 기본적인 데이터 몇개(5)를 확인해 준다.
# [데이터].isnull().sum() : 각 열의 결측치(비어 있는 값)의 개수
```

```python
# 요일별 식사 금액은 얼마나 될까?
sns.barplot(x='day', y='total_bill', data=tips)
plt.title("Amount of meal")
```

```python
sns.boxplot(x='day', y='total_bill', data=tips)
plt.title("Amount of meal")
```

```python
plt.figure(figsize=(8,6))
sns.barplot(x='day', y='total_bill', hue='smoker', data=tips)
plt.title("total bills versus days")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.barplot(x='day', y='total_bill', hue='sex', data=tips)
plt.title("total bills versus days")
plt.show()
```

```python
fg = sns.load_dataset('flights')
fg.head()
```

```python
### info(), plt.plot(tips.tip[0:10], tips.total_bill[0:10], 'o')
plt.figure(figsize=(16,8))
sns.barplot(x='month', y='passengers', hue='year', data=fg)
plt.title("Numberof passengers")
plt.show()
```

```python
sns.scatterplot(x='year', y='passengers', data=fg)
```

```python
fgpivot = fg.pivot("month", "year", "passengers")
fgpivot
plt.figure(figsize=(15,10))
sns.heatmap(fgpivot)
```
